##### Data Description - Table transaction:

- TransactionDT: différence de temps (timedelta) par rapport à une date de référence donnée (ce n’est pas un horodatage réel).

- isFraud : étiquette

- TransactionAMT : montant du paiement de la transaction, en dollars américains (USD)

- ProductCD : code produit, identifiant le produit pour chaque transaction

- card1 - card6 : informations sur la carte de paiement, comme le type de carte, la catégorie, la banque émettrice, le pays, etc.

- addr : adresse

- dist : distance approximative entre l’adresse de la personne et l’adresse du commerçant

- P_ et R__ emaildomain : domaine d’adresse e-mail de l’acheteur (P_) et du destinataire (R__)

- C1 - C14 : variables de comptage, par exemple combien d’adresses sont associées à la carte de paiement, etc. (la signification exacte est masquée)

- D1 - D15 : différences de temps (timedelta), par exemple le nombre de jours entre deux transactions, nombre de jours depuis la création du compte, etc.

- M1 - M9 : correspondances (match), par exemple entre le nom sur la carte et l’adresse, etc.

- Vxxx : caractéristiques enrichies créées par Vesta, incluant des classements, des comptages et d’autres relations entre entités.

```python
from multiprocessing import cpu_count
from pathos.multiprocessing import ProcessingPool as Pool
import numpy as np
from collections import defaultdict

def _missing_pattern_key(args):
    col, col_mask = args
    return (col, tuple(col_mask)) 

def find_always_missing_together(df, verbose=True):
    """
    Trouve les colonnes qui ont exactement le même pattern de valeurs manquantes
    sur TOUT le dataset.

    Paramètres:
    -----------
    df : DataFrame à analyser (complet, pas d'échantillon)
    verbose : bool, default=True pour Afficher les informations de progression
    
    Returns:
    --------
    list : Liste de listes, chaque sous-liste contient les colonnes qui 
           manquent toujours ensemble
    """

    if verbose:
        print("="*70)
        print("🔍 ANALYSE DES PATTERNS DE VALEURS MANQUANTES")
        print("="*70)
        print(f"Taille du dataset : {df.shape[0]:,} lignes et {df.shape[1]} colonnes")

    # 1. Créer le masque de valeurs manquantes
    if verbose:
        print("\n⏳ Étape 1/4 : Création du masque de NaN...")

    mask = df.isna()

    # 2. Filtrer les colonnes pertinentes
    if verbose:
        print("⏳ Étape 2/4 : Filtrage des colonnes...")

    # Garde seulement les colonnes avec au moins 1 NaN et pas 100% NaN
    mask = mask.loc[:, (mask.sum() >0) & (mask.sum() < len(mask))]

    cols = mask.columns

    if verbose: 
        print(f"   ✓ {len(cols)} colonnes avec valeurs manquantes à analyser")
        print(f"   ✓ {df.shape[1] - len(cols)} colonnes ignorées (0% ou 100% NaN)")

    if len(cols) == 0:
        if verbose:
            print("\n✅ Aucune colonne avec valeurs manquantes partielles !")
        return []

    # 3. Traitement Parallèle : répartir les colonnes sur plusieurs processus
    if verbose:
        print(f"\n⏳ Étape 3/4 : Analyse parallèle avec {cpu_count()} CPU...")
        print(f"   (Cela peut prendre quelques minutes sur gros dataset...)")
 
    with Pool(cpu_count()) as pool:
        col_patterns = pool.map(
            _missing_pattern_key,
            [(col, mask[col].values) for col in cols]
        )

    if verbose:
        print("   ✓ Analyse terminée !")

    # 4. Regrouper les colonnes par pattern identique
    if verbose:
        print("\n⏳ Étape 4/4 : Regroupement des colonnes...")

    pattern_groups = defaultdict(list)
    for col, pattern in col_patterns:
        pattern_groups[pattern].append(col)

    # Garder seulement les groupes avec 2+ colonnes
    always_together = [
        group 
        for group in pattern_groups.values() 
        if len(group) > 1
    ]

    # 5. Afficher les résultats
    if verbose:
        print("\n" + "="*70)
        print("📋 RÉSULTATS")
        print("="*70)
        
        if not always_together:
            print("✅ Aucune colonne ne manque systématiquement ensemble")
            print("   → Toutes les colonnes ont des patterns de NaN uniques")
        else:
            print(f"🎯 {len(always_together)} groupe(s) de colonnes trouvé(s) :\n")
            
            for i, group in enumerate(always_together, 1):
                print(f"📌 Groupe {i} ({len(group)} colonnes) :")
                print(f"   {group}")
                
                # Calculer des stats sur ce groupe
                num_missing = mask[group[0]].sum()  # Toutes ont le même pattern
                pct_missing = (num_missing / len(df)) * 100
                print(f"   └─ {num_missing:,} lignes manquantes ({pct_missing:.2f}%)")
                print()
        
        print("="*70)
    
    return always_together

always_missing_together = find_always_missing_together(df_train)
always_missing_together
    
```

```python
# transformer chaque valeur de texte en un nombre unique
import json

def alphabetic_label_encoding(df, categorical_cols):
    df_encoded = df.copy()

    # Dictionnaire pour enregistrer les correspondances
    mappings = {} # {"colonne1": {"A": 1, "B": 2, "C": 3}, "colonne2": {"A": 1, "B": 2, "C": 3}}

    for col in categorical_cols:
        categories = sorted(df_encoded[col].dropna().unique())
        mapping = {cat: i+1 for i, cat in enumerate(categories)} 

        mappings[col] = mapping
        df_encoded[col] = df_encoded[col].map(mapping)

    with open('label_mappings.json', 'w') as f:
        json.dump(mappings, f, indent=2)

    return df_encoded

df_train_encoded = alphabetic_label_encoding(df_train, categorical_cols)

        
```

- Pour les colonnes qui ont des valeurs manquantes ensemble on les réduit en une seule colonne par groupe.

```python
from sklearn.cluster import FeatureAgglomeration
import numpy as np

def agglomerate_missing_groups_robust(df, groups, categorical_cols, n_clusters=1):
    df_new = df.copy()
    for i, group in enumerate(groups, 1):
        numeric_group = [col for col in group if col not in categorical_cols and col in df_new.columns]

        # pas de colonnes numériques -> skip
        if not numeric_group:
            continue

        # retirer colonnes entièrement NaN
        numeric_group = [col for col in numeric_group if not df_new[col].isna().all()]
        if not numeric_group:
            # toutes les colonnes étaient vides -> créer col agrégée vide et drop originals
            df_new[f"A{i}"] = np.nan
            df_new = df_new.drop(columns=[c for c in group if c in df_new.columns], errors='ignore')
            continue

        # mask : lignes où au moins une colonne a une valeur
        mask = ~df_new[numeric_group].isna().all(axis=1)
        if not mask.any():
            df_new[f"A{i}"] = np.nan
            df_new = df_new.drop(columns=numeric_group)
            continue

        # cas 1 colonne -> on la recopie (pas d'agglomération possible)
        if len(numeric_group) == 1:
            col = numeric_group[0]
            df_new[f"A{i}"] = np.nan
            df_new.loc[mask, f"A{i}"] = df_new.loc[mask, col]
            df_new = df_new.drop(columns=[col])
            continue

        # >=2 colonnes : préparer X
        col_means = df_new[numeric_group].mean()
        X = df_new.loc[mask, numeric_group].fillna(col_means)

        # s'assurer que n_clusters <= n_features
        use_clusters = max(1, min(n_clusters, len(numeric_group)))
        if use_clusters == 1:
            # si on demande 1 cluster, on peut juste faire la moyenne/PC1
            df_new[f"A{i}"] = np.nan
            df_new.loc[mask, f"A{i}"] = X.mean(axis=1)
        else:
            agg = FeatureAgglomeration(n_clusters=use_clusters)
            reduced = agg.fit_transform(X)  # shape = (n_rows_mask, use_clusters)
            if reduced.ndim == 1 or reduced.shape[1] == 1:
                df_new[f"A{i}"] = np.nan
                df_new.loc[mask, f"A{i}"] = reduced.ravel()
            else:
                # créer A{i}_0, A{i}_1, ...
                for k in range(reduced.shape[1]):
                    col_name = f"A{i}_{k}"
                    df_new[col_name] = np.nan
                    df_new.loc[mask, col_name] = reduced[:, k]

        # supprimer les colonnes originales du groupe
        df_new = df_new.drop(columns=numeric_group)

    return df_new

# Utilisation
df_train_reduced = agglomerate_missing_groups_robust(df_train_encoded, always_missing_together, categorical_cols, n_clusters=1)
df_train_reduced.head()
```

```python
# les NaN vont être remplacé par 0, donc _nan est utile Pour savoir si la veleur était NaN
    # 1 -> NaN
    # 0 -> 0

def add_nan_indicators_and_convert(df):
    df_new = df.copy()
    nan_indicators = {}

    for col in df_new.columns:
        if col in ["TransactionID", "isFraud"] or col.endswith("_nan"):
            continue
        nan_indicators[f"{col}_nan"] = df_new[col].isna().astype(int)
        indicators_df = pd.DataFrame(nan_indicators, index=df_new.index)

    cols = []
    for col in df_new.columns:
        cols.append(col)
        if f"{col}_nan" in indicators_df.columns:
            cols.append(f"{col}_nan") # ordre final: [col1, col1_nan, col2, col2_nan, ...]

    combined = pd.concat([df_new, indicators_df], axis=1)
    combined = combined[cols] # rêordonner les colonnes

    for col in combined.columns:
        combined[col] = pd.to_numeric(combined[col], errors='coerce')
    
    combined = combined.fillna(0)
    return combined

df_train_reduced = add_nan_indicators_and_convert(df_train_reduced)
df_train_reduced.head()
    
    
```

- On remarque dans le dataset que certaines colonnes _nan sont toujours à 0. Ces colonnes là indiquent que les colonnes d’origine n’ont jamais eu de NaN.

```python
# exemple :
df_train_reduced['C13_nan'].unique() # toutes les valeurs sont pas NaN
```

- Ces colonnes loi n’apportent aucune information au modèle (pas de variance → corrélation NaN).

- Elles augmentent inutilement la dimensionnalité (bruit + temps d'entraînement).

- Elles n'aident pas à détecter la fraude, une colonne constante (toujours 0) ne peut pas aider à prédire quoi que ce soit

```python
# on vérifie si il existe des colonnes _nan constantes avec toujours 1
always_one_nan_cols = [
    col for col in df_train_reduced.columns
    if col.endswith('_nan') and df_train_reduced[col].nunique() == 1 and df_train_reduced[col].iloc[0] == 1
]

print(f"{len(always_one_nan_cols)} colonnes _nan = 1 partout : {always_one_nan_cols}")
```

```python
# Donc on supprime les colonnes _nan constantes avec toujours 0
    # Ça voudrait aussi dire que si une colonne _nan existe c'est que :
    # - La colonne d'origine a des valeurs NaN 

def find_always_zero_nan_indicators(df):
    df_new = df.copy()
    return [
        col for col in df_new.columns
        if col.endswith('_nan') and df_new[col].nunique() == 1 
    ]

always_zero_nan_cols = find_always_zero_nan_indicators(df_train_reduced)

def remove_always_zero_nan_indicators(df, always_zero_nan_cols):
    df_new = df.copy()
    print(f"{len(always_zero_nan_cols)} colonnes _nan constantes supprimées : {always_zero_nan_cols}")

    if always_zero_nan_cols:
        df_new.drop(columns=always_zero_nan_cols, inplace=True)
    
    return df_new

df_train_final = remove_always_zero_nan_indicators(df_train_reduced, always_zero_nan_cols)
df_train_final.head()



```

- Analyser la distribution de chaque variable numérique
- Repérer celles qui sont trop asymétriques ou non normales.

```python
from scipy.stats import normaltest, skew

def normaltest_and_skew_report(df):
    results = []
    skip_cols = {"TransactionID", "isFraud"}
    numeric_cols = [col for col in df.select_dtypes(include='number') if col not in skip_cols and not col.endswith('_nan')]
    
    for col in numeric_cols:
        nan_indicator_col = f"{col}_nan"

        # utiliser SEULEMENT les lignes où indicator est 0 OU la colonne indicator n'existe pas
        if nan_indicator_col in df.columns:
            data = df.loc[df[nan_indicator_col] == 0,col].dropna() # on veut séléctionner les valeurs de la colonne où son indicator est 0 (valeur réelle)
        else:
            data = df[col].dropna()
        
        # Cas : pas assez pour faire des tests statistiques
        if len(data) < 8:
            continue

        stat, p = normaltest(data)
        col_skew = skew(data)

        results.append({
            'column' : col,
            'normal_stat' : stat,
            'normal_p' : p,
            'skewness' : col_skew,
        })

    return pd.DataFrame(results)

report = normaltest_and_skew_report(df_train_final)
report
```

- p > 0.05 : On ne rejette pas l’hypothèse de normalité → la colonne est plutôt normale

- p ≤ 0.05 : On rejette l’hypothèse de normalité → la colonne n’est pas normale (asymétrique, biaisée, etc.)

```python
def get_normal_columns(report_df, alpha=0.05):
    """
    Returns a list of columns that pass the normality test (p > alpha).
    """
    return report_df.loc[report_df['normal_p']> alpha, 'column'].tolist()

normal_cols = get_normal_columns(report)
normal_cols

```
